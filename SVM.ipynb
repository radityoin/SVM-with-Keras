```
import numpy as np

#Library SVM
from keras.regularizers import l2
from keras.models import Sequential
from keras.layers import Dense

#Library image
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```
#Implementasi sklearn.svm SVC classifier menggunakan Keras
#source code: https://stackoverflow.com/questions/54414392/convert-sklearn-svm-svc-classifier-to-keras-implementation

classifier = Sequential()
classifier.add(Dense(64, activation='relu'))
classifier.add(Dense(1,activation='linear', W_regularizer=l2(0.01)))
classifier.compile(loss='hinge', optimizer='adadelta', metrics=['accuracy'])
```

```
#Mencocokan SVM dengan gambar

#Data training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range = 0.15,
    height_shift_range = 0.15,
    shear_range=0.01,
    zoom_range=0.1,
    horizontal_flip=True)
training_set =  train_datagen.flow_from_directory(
    directory = '/content/drive/My Drive/TVST/train',
    target_size = (64,64),
    color_mode = 'rgb',
    batch_size = 32,
    class_mode = 'binary',
    shuffle = True
    )

#Data test
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
    directory = '/content/drive/My Drive/TVST/test inna',
    target_size = (64,64),
    color_mode = 'rgb',
    batch_size = 32,
    class_mode = 'binary',
    shuffle = False
    )
```

```
#Mengetes gambar gedung pada SVM yang telah dibuat
test_image = image.load_img('/content/drive/My Drive/TVST/COBA/Perpustakaan.jpg', target_size = (64,64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = classifier.predict(test_image)
training_set.class_indices
nilai = (abs(np.mean(result [0][0])))
if  nilai >= 50:
  prediction = 'TVST'
else:
  prediction = 'Bukan TVST'
print(prediction)
print(nilai,'%')
```

```
classifier.compile(loss='hinge', optimizer='adadelta', metrics=['accuracy'])

history = classifier.fit_generator(
        training_set,
        steps_per_epoch=4,
        epochs=20,
        validation_data=test_set,
        validation_steps=2
        )

loss_train = history.history['loss']
loss_val = history.history['val_loss']
acc_train = history.history['acc']
acc_val = history.history['val_acc']
```
